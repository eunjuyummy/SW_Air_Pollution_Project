## 일산화 탄소 범위 수정

```python
# 데이터 로드
data = pd.read_csv('/content/drive/MyDrive/Datasets/seoul/seoul_data.csv')

# 대기지수 점수를 계산하는 함수
def calculate_aqi(Cp, BPLO, BPHI, ILO, IHI):
    return ((IHI - ILO) / (BPHI - BPLO)) * (Cp - BPLO) + ILO

# 대기오염물질의 농도 구간 값
ozone_params = [(0, 0.03, 0, 50), (0.031, 0.09, 51, 100), (0.091, 0.15, 101, 250), (0.151, 0.6, 251, 500)] # 오존
pm25_params = [(0, 15, 0, 50), (16, 35, 51, 100), (36, 75, 101, 250), (76, 500, 251, 500)] # 초미세먼지
so2_params = [(0, 0.02, 0, 50), (0.021, 0.05, 51, 100), (0.051, 0.15, 101, 250), (0.151, 1, 251, 500)] # 아황산가스
co_params = [(0, 2, 0, 50), (2.01, 9, 51, 100), (9.01, 15, 101, 250), (15.01, 50, 251, 500)] # 일산화탄소
no2_params = [(0, 0.03, 0, 50), (0.031, 0.06, 51, 100), (0.061, 0.2, 101, 250), (0.201, 2, 251, 500)] # 이산화질소
pm10_params = [(0, 30, 0, 50), (31, 80, 51, 100), (81, 150, 101, 250), (151, 600, 251, 500)] # 미세먼지

# 각 오염물질별 대기지수 계산
def get_aqi_value(value, params):
    for BPLO, BPHI, ILO, IHI in params:
        if BPLO <= value <= BPHI:
            return calculate_aqi(value, BPLO, BPHI, ILO, IHI)
    return None

# 각 오염물질에 대해 대기지수 계산
data['Ozone_CAI'] = data['Ozone'].apply(lambda x: get_aqi_value(x, ozone_params)) # 오존 배출량
data['PM2.5_CAI'] = data['PM2.5'].apply(lambda x: get_aqi_value(x, pm25_params)) # 초미세먼지 배출량
data['SO2_CAI'] = data['SO2'].apply(lambda x: get_aqi_value(x, so2_params)) # 아황산가스 배출량
data['CO_CAI'] = data['CO'].apply(lambda x: get_aqi_value(x, co_params)) # 일산화탄소 배출량
data['NO2_CAI'] = data['NO2'].apply(lambda x: get_aqi_value(x, no2_params)) # 질소산화물 배출량
data['PM10_CAI'] = data['ALL_PM'].apply(lambda x: get_aqi_value(x, pm10_params)) # 미세먼지 배출량

# 통합대기환경지수 계산
def calculate_overall_aqi(row):
    aqi_values = [row['Ozone_CAI'], row['PM2.5_CAI'], row['SO2_CAI'], row['CO_CAI'], row['NO2_CAI'], row['PM10_CAI']]
    # None 값을 제외하고 계산
    aqi_values = [value for value in aqi_values if value is not None]
    if not aqi_values:  # 모든 값이 None인 경우
        return None
    max_aqi = max(aqi_values)
    num_bad_aqi = sum(1 for value in aqi_values if value >= 101)  # '나쁨' 이상의 등급 개수

    if num_bad_aqi == 1:
        return max_aqi
    elif num_bad_aqi == 2:
        return max_aqi + 50
    elif num_bad_aqi >= 3:
        return max_aqi + 75
    return max_aqi

data['CAI'] = data.apply(calculate_overall_aqi, axis=1)

# NaN 값을 각 숫자형 칼럼의 평균값으로 대체
numeric_columns = data.select_dtypes(include=['float64', 'int64']).columns
data[numeric_columns] = data[numeric_columns].fillna(data[numeric_columns].mean())

# 새로운 데이터프레임에 포함할 열 목록
columns_to_select = ['sigungu_code', 'year', 'Ozone_CAI', 'PM2.5_CAI', 'SO2_CAI', 'CO_CAI', 'NO2_CAI', 'PM10_CAI', 'CAI']

# 새로운 데이터프레임 생성
selected_data = data[columns_to_select]

# 선택된 데이터를 CSV 파일로 저장
selected_data.to_csv('/content/drive/MyDrive/Datasets/seoul/CAI_seoul_data.csv', index=False)
selected_data
```

# 그래프 생성

```python
import pandas as pd
import matplotlib.pyplot as plt
import random

data = pd.read_csv('/content/drive/MyDrive/follium/Data/Graph_CAI_seoul.csv')

columns_to_plot = ['CAI']

colors = ['orange', 'blue', 'green', 'red', 'purple', 'brown', 'pink', 'gray', 'olive']

unique_sigungu_codes = data['sigungu_code'].unique()

for sigungu_code in unique_sigungu_codes:
    df_sigungu = data[data['sigungu_code'] == sigungu_code]
    plt.figure(figsize=(10, 5))

    color = random.choice(colors)

    plt.plot(df_sigungu['year'], df_sigungu['CAI'], marker='o', color=color, linewidth=5)
    plt.xlabel('Year', fontsize=23)
    plt.ylabel('CAI', fontsize=23)
    plt.xticks(df_sigungu['year'].unique(), fontsize=12)
    plt.yticks(fontsize=20)

    # Set absolute tight y-axis limits based on the data range
    y_min = df_sigungu['CAI'].min()
    y_max = df_sigungu['CAI'].max()
    plt.ylim(y_min, y_max)

    plt.tight_layout(pad=2.0)
    plt.savefig(f'/content/drive/MyDrive/follium/img/seoul/{sigungu_code}.png')
    plt.close()
```
